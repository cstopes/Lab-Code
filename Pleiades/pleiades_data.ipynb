```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sp
```

```python
data = pd.read_csv("pleiades-extended-dataset.csv")
```

```python
print(data)
```

```python
v = data.loc[:,"v"]
b = data.loc[:,"b"]
u = data.loc[:,"u"]

bv = b-v
print(bv)
```

```python
bvabsolute = np.array([-0.35, -0.31, -0.16, 0, 0.13, 0.27, 0.42, 0.58, 0.7, 0.89, 1.18, 1.45, 1.63, 1.8])
absolutemags = np.array([-5.8, -4.1, -1.1, -0.7, 2, 2.6, 3.4, 4.4, 5.1, 5.9, 7.3, 9.0, 11.8, 16])
```

```python
fig, ax = plt.subplots()

Doffset = 5.5
offsetuncertainty = 0.5
Doffset_plus_err = Doffset + offsetuncertainty
Doffset_less_err = Doffset - offsetuncertainty

ax.plot(bvabsolute, absolutemags + (Doffset_less_err), 'k:', label='Uncertainty')
ax.plot(bvabsolute, absolutemags + (Doffset), 'b', label='Offset HRD Model')
ax.plot(bvabsolute, absolutemags + (Doffset_plus_err), 'k:')

ax.scatter(bv, v, marker="+", color='red', label='Pleiades Stars')

ax.set_xlim([-0.2, 1.75])
ax.set_ylim([20,0])
ax.set_xlabel("Colour Index (B - V)")
ax.set_ylabel("Apparent Magnitude m_V")
ax.spines['left'].set_color("red")
ax.spines['right'].set_color("blue")

secax = ax.secondary_yaxis('right')
secax.set_ylabel('Absolute Magnitude M_V + %.1f' %Doffset)
secax.set_ylim([6, 26])
secax.spines['left'].set_color("blue")

plt.legend()
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots()

Doffset = 5.5
offsetuncertainty = 0.5
Doffset_plus_err = Doffset + offsetuncertainty
Doffset_less_err = Doffset - offsetuncertainty

ax.plot(bvabsolute, absolutemags, 'b', label='Model of Hertzsprung-Russell Diagram')

ax.scatter(bv, v, marker="+", color='red', label='Pleiades Stars')

ax.set_xlim([-0.2, 1.75])
ax.set_ylim([20,0])
ax.set_xlabel("Colour Index (B - V)")
ax.set_ylabel("Apparent Magnitude m_V")
ax.spines['left'].set_color("red")
ax.spines['right'].set_color("blue")

secax = ax.secondary_yaxis('right')
secax.set_ylabel('Absolute Magnitude M_V + %.1f' %Doffset)
secax.set_ylim([6, 26])
secax.spines['left'].set_color("blue")

plt.legend()
plt.grid()
plt.show()
```

```python
def d(distanceoffset):
    return 10**((distanceoffset + 5)/5)

D= d(Doffset)

Dpluserr, Dminuserr = d(Doffset_plus_err), d(Doffset_less_err)

Derr1 = Dpluserr - D
Derr2 = D - Dminuserr
Derr = np.maximum(Derr1, Derr2)

print(D)
print(Derr)
```
